```python
from jinja2 import *
```

```python
parameters=[
    {
        "table":"etl_cata.silver.factstream",
        "alias":"fs",
        "cols":"fs.stream_id,fs.listen_duration"
    },
    {
        "table":"etl_cata.silver.dimuser",
        "alias":"user",
        "cols":"user.user_id,user.user_name",
        "condition":"fs.user_id=user.user_id"
    },
    {
        "table":"etl_cata.silver.dimtrack",
        "alias":"track",
        "cols":"track.track_id,track.track_name",
        "condition":"fs.track_id=track.track_id"
    }

]
```

```python
query="""
         SELECT
	{% for params in parameter%}
        {{params.cols}}
	    {% if not loop.last %}
		    ,
        {% endif %}
    {% endfor %}
    FROM
    {% for params in parameter%}
        {% if loop.first %}
            {{params['table']}} as {{params['alias']}}
        {% endif %}
        {% if not loop.first %}
            LEFT JOIN
            {{params['table']}} as {{params['alias']}}
            ON {{params['condition']}}
        {% endif %}
    {% endfor %}            		
"""
```

```python
jinja_sql_query=Template(query)
sql_query=jinja_sql_query.render(parameter=parameters)
print(sql_query)
```

```python
display(spark.sql(sql_query))
```
